```python
import os

# Define the base directory for the stock data
base_dir = r"D:\Kifya_training\Week 1\Technical  Content\Data\yfinance_data"

# Define file paths for each stock with correct filenames
file_names = [
    "AAPL_historical_data.csv",
    "AMZN_historical_data.csv",
    "GOOG_historical_data.csv",
    "META_historical_data.csv",
    "MSFT_historical_data.csv",
    "NVDA_historical_data.csv",
    "TSLA_historical_data.csv"
]

required_columns = ['Open', 'High', 'Low', 'Close', 'Volume']

def load_data(file_path):
    return pd.read_csv(file_path)

for file_name in file_names:
    file_path = os.path.join(base_dir, file_name)
    if os.path.exists(file_path):
        df = load_data(file_path)
        missing_columns = [col for col in required_columns if col not in df.columns]
        if not missing_columns:
            print(f"The dataset {file_name} contains all required columns.")
        else:
            print(f"The dataset {file_name} is missing columns: {missing_columns}")
    else:
        print(f"Error: The file at {file_path} does not exist.")
```

```python

# Function to calculate financial metrics
def calculate_metrics(df):
    # Calculate daily returns
    returns = df['Adj Close'].pct_change()

    # Calculate daily volatility
    volatility = returns.std()

    # Annualize the volatility (assuming 252 trading days in a year)
    annual_volatility = volatility * (252 ** 0.5)

    # Calculate the average daily return
    avg_daily_return = returns.mean()

    # Assume a risk-free rate (e.g., 0.01 for 1%)
    risk_free_rate = 0.01

    # Calculate the daily excess return
    excess_return = avg_daily_return - risk_free_rate / 252

    # Calculate the Sharpe Ratio
    sharpe_ratio = excess_return / volatility

    # Annualize the Sharpe Ratio
    annual_sharpe_ratio = sharpe_ratio * (252 ** 0.5)

    return annual_volatility, annual_sharpe_ratio

# Loop through each file, apply metrics calculation, and display results
for file_name in file_names:
    file_path = os.path.join(base_dir, file_name)
    if os.path.exists(file_path):
        df = load_data(file_path)
        annual_volatility, annual_sharpe_ratio = calculate_metrics(df)
        print(f"\nFinancial metrics for {file_name}:")
        print(f"Annualized Volatility: {annual_volatility}")
        print(f"Annualized Sharpe Ratio: {annual_sharpe_ratio}")
    else:
        print(f"Error: The file at {file_path} does not exist.")
```
